## ENCODING ALL CATEGORICAL VARIABLES
strongly disagree < disagree < neutral < agree < strongly agree

```python
df_model = df_model.copy()
```

```python
ordinal_cols = [
    'Ease and convenient',
    'Late Delivery',
    'Bad past experience',
    'More Offers and Discount'
]

binary_cols = ['Influence of rating']

for col in ordinal_cols + binary_cols:
    df_model[col] = (
        df_model[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )
```

```python
ordinal_map = {
    'strongly disagree': 1,
    'disagree': 2,
    'neutral': 3,
    'agree': 4,
    'strongly agree': 5
}


binary_map = {
    'yes': 1,
    'no': 0
}
```

```python
# Ordinal encoding
for col in ordinal_cols:
    df_model[col] = df_model[col].map(ordinal_map)

# Binary encoding
df_model['Influence of rating'] = df_model['Influence of rating'].map(binary_map)
```

```python
# Fill ordinal with median
for col in ordinal_cols:
    df_model[col] = df_model[col].fillna(df_model[col].median())

# Fill binary with mode
df_model['Influence of rating'] = df_model['Influence of rating'].fillna(
    df_model['Influence of rating'].mode()[0]
)
```

```python
df_model[ordinal_cols + binary_cols].isnull().sum()
```

```python
df_model.head(5)
```

```python
nominal_cols = [
    'Occupation',
    'Frequently used Medium',
    'Frequently ordered Meal category ',
    'Perference'
]

df_model = pd.get_dummies(
    df_model,
    columns=nominal_cols,
    drop_first=True
)
```

## Combining two columns to create target column for our model

```python
## (column1) discount sensitive columns 
df_model['is_discount_sensitive'] = (df_model['More Offers and Discount'] >= 4).astype(int)

##(column2) frequency of orders without discounts
high_order_threshold = df_model['No. of orders placed'].quantile(0.7)

df_model['is_high_order'] = (df_model['No. of orders placed'] >= high_order_threshold).astype(int)
```

```python
df_model['target_disocunt_growth']= (
(df_model['is_discount_sensitive']==1) & 
(df_model['is_high_order']==0)).astype(int)
```
